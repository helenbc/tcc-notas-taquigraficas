```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def calcular_metricas_com_mapping(predicoes, ground_truth, mapping):
    # Inicializar listas para as predições e os rótulos verdadeiros
    predicoes_lista = []
    ground_truth_lista = []
    
    # Iterar sobre as predições e os rótulos verdadeiros
    for topico, categoria in mapping.items():
        # Adicionar as predições e os rótulos verdadeiros para o tópico atual às listas
        predicoes_lista.append(predicoes[topico])
        ground_truth_lista.append(ground_truth[topico])
    
    # Calcular métricas
    acuracia = accuracy_score(ground_truth_lista, predicoes_lista)
    precisao = precision_score(ground_truth_lista, predicoes_lista, average='macro')
    recall = recall_score(ground_truth_lista, predicoes_lista, average='macro')
    f1 = f1_score(ground_truth_lista, predicoes_lista, average='macro')
    
    return acuracia, precisao, recall, f1
```
